```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

study_hours = np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)

marks = np.array([35,42,50,58,65,72,82,90])

model = LinearRegression()

model.fit(study_hours, marks)

prediction = model.predict([[6]])

print("Predicted Marks:", prediction[0])

plt.figure(figsize=(8,5))

plt.scatter(study_hours, marks, color="blue", label="Actual Data")

plt.plot(
study_hours,
model.predict(study_hours),
color="red",
linewidth=2,
label="Regression Line"

)

plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.title("Study Hours vs Marks")

plt.legend()

plt.grid(False)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

study_hours = np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)

marks = np.array([35,42,50,58,65,72,182,190])

model = LinearRegression()

model.fit(study_hours, marks)

prediction = model.predict([[6]])

print("Predicted Marks:", prediction[0])

plt.figure(figsize=(8,5))

plt.scatter(study_hours, marks, color="blue", label="Actual Data")

plt.plot(
study_hours,
model.predict(study_hours),
color="red",
linewidth=2,
label="Regression Line"

)

plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.title("Study Hours vs Marks")

plt.legend()

plt.grid(False)

plt.show()
```

```python
predicted = model.predict(study_hours)

comparison = pd.DataFrame({

"study hours":study_hours.flatten(),
"actual marks" : marks,
"predicted marks":predicted.round(2)

})

comparison["error"] = abs (
comparison["actual marks"] - comparison["predicted marks"]

)

print(comparison)
```
